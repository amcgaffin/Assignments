11- Merge these two DataFrames into a single DataFrame called census.
Please use the following dataset for merging purposes:

'https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/state-populations.csv'

'https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/census-divisions.csv'

```python
pd.read_csv('census-divisions.csv')
pd.read_csv('state-populations.csv')
census_div = pd.read_csv('census-divisions.csv')
census_pop = pd.read_csv('state-populations.csv')
census = pd.merge(census_pop, census_div, on = 'state',how='inner')
census.head()
```

12- Re-shape census so that one column contains all population measures, and another the year attributes.


```python
reshaped_census = census_pop.melt(id_vars='state', var_name='year', value_name='population')
reshaped_census.head()
```

13- Group the data by year and summarize it.


```python
yearly_sum = reshaped_census.groupby('year')['population'].sum().reset_index()
yearly_sum
```

14- Group the data by region, division and year and summarize it.


```python
merged_census_pop = pd.merge(reshaped_census, census_div, on='state')
grouped_summary = merged_census_pop.groupby(['region', 'division', 'year'])['population'].sum().reset_index()
grouped_summary
```

15- Create a crosstab between state and category showing amount pledged as the value.
Please refer to the following link for this practice:

https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/ksprojects.csv

```python
ks = pd.read_csv("ksprojects.csv")
crosstab = pd.crosstab(index=ks['state'],
                       columns=ks['category'],
                       values=ks['usd_pledged'],
                       aggfunc='sum',
                       dropna=True)
crosstab
```
